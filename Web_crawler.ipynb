### Step1:获取doi列表

```python
import os
import re
import pandas as pd
from crossref_commons.iteration import iterate_publications_as_json
from Web_crawler.data_archive import data_archive
```

```python
# Elsevier:10.1016
filters = {'prefix': '10.1016', 'type': 'journal-article'}     
pattern = re.compile(r'single[- ]crystal|monocrystal|monocrystalline|single[- ]crystalline', re.IGNORECASE)
re.compile(r'(single[- ]crystal|monocrystal|monocrystalline|single[- ]crystalline).*alloy', re.IGNORECASE)
queries = {'query.bibliographic': pattern}
doilist_elsevier = []     
for p in iterate_publications_as_json(max_results=10, filter=filters, queries=queries):
    doilist_elsevier.append(p['DOI'])
    print(p['DOI'])
```

```python
doilist_elsevier
```

```python
# 存入表格
doi_path = r"./elsevier_doi.xlsx"
data_archive(doilist_elsevier, doi_path, "DOI")
```
